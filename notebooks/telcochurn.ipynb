```python
# Read the main DataFrame

df = pd.read_excel(r'./../data/Telco_customer_churn.xlsx')
```

```python
# Read auxiliar DataFrames

df_demographics = pd.read_excel(r'./../data/Telco_customer_churn_demographics.xlsx')
df_location = pd.read_excel(r'./../data/Telco_customer_churn_location.xlsx')
df_population = pd.read_excel(r'./../data/Telco_customer_churn_population.xlsx')
df_services = pd.read_excel(r'./../data/Telco_customer_churn_services.xlsx')
df_status = pd.read_excel(r'./../data/Telco_customer_churn_status.xlsx')
```

```python
# Creating list of auxiliar DataFrames to iterate

aux_dfs = [df_demographics, df_location, df_population, df_services, df_status]
aux_dfs_names = ['df_demographics', 'df_location', 'df_population', 'df_services', 'df_status']
```

```python
# df columns, info and sample

print(df.columns, '\n')
print(df.info(), '\n')
with pd.option_context('display.max_columns', 33):
    display(df.sample(5))
```

```python
for i, d in enumerate(aux_dfs):
    print(aux_dfs_names[i], ':\n')
    print(d.columns, '\n')
    print(d.info(), '\n')
```

```python
# Looking how many unique values are in each column

for c in df.columns:
    print(f'Cardinalidad de {c}: {df[c].nunique()}')
```

```python
# Unique value for columns that have only one

for c in df.columns:
    if df[c].nunique() == 1:
        print(df[c].value_counts(), '\n')
```

```python
for i, d in enumerate(aux_dfs):
    print(aux_dfs_names[i], ':\n')
    for c in d.columns:
        print(f'Cardinalidad de {c}: {d[c].nunique()}')
    print('\n')
```

```python
for i, d in enumerate(aux_dfs):
    print(aux_dfs_names[i], ':\n')
    for c in d.columns:
        if d[c].nunique() == 1:
            print(df[c].value_counts(), '\n')
    print('\n')
```

```python
# Looking for columns in auxiliar DataFrames not included in the main one
# Some could be the same columns with different names

for i, d in enumerate(aux_dfs):
    print(aux_dfs_names[i], ':\n')
    for c in d.columns:
        if c not in df.columns:
            print(f'Column {c} is not in main DataFrame')
    print('\n')
```
